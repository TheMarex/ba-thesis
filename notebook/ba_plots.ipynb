```python
%pylab inline

import pandas as pd
import matplotlib.pyplot as plt

import utils

size = (18, 6)

# Finds newest log and calls pd.read_csv(..)
index_col = "Timestamp"

class LogData:
    def __init__(self, log_dir, n):
        com_data_path         = utils.get_nth_newest(log_dir, "com.csv", n)
        zmp_data_path         = utils.get_nth_newest(log_dir, "zmp.csv", n)
        trajectory_data_path  = utils.get_nth_newest(log_dir, "trajectory.csv", n)
        support_data_path     = utils.get_nth_newest(log_dir, "support.csv", n)
        collision_data_path   = utils.get_nth_newest(log_dir, "collision.csv", n)
        cp_data_path          = utils.get_nth_newest(log_dir, "cp.csv", n)
        robot_data_path       = utils.get_nth_newest(log_dir, "robot.csv", n)
        ankle_data_path       = utils.get_nth_newest(log_dir, "ankle.csv", n)

        print("""
        Using the following logs:
        com:         %s
        zmp:         %s
        trajectory:  %s
        collision:   %s
        support:     %s
        cp:          %s
        robot:       %s
        ankle:       %s
        """ % (com_data_path,
               zmp_data_path,
               trajectory_data_path,
               collision_data_path,
               support_data_path,
               cp_data_path,
               robot_data_path,
               ankle_data_path))

        self.com_data         = pd.read_csv(com_data_path, index_col=index_col)
        self.zmp_data         = pd.read_csv(zmp_data_path, index_col=index_col)
        self.trajectory_data  = pd.read_csv(trajectory_data_path, index_col=index_col)
        self.collision_data   = pd.read_csv(collision_data_path, index_col=index_col)
        self.support_data     = pd.read_csv(support_data_path, index_col=index_col)
        self.cp_data          = pd.read_csv(cp_data_path, index_col=index_col)
        self.robot_data       = pd.read_csv(robot_data_path, index_col=index_col)
        self.ankle_data       = pd.read_csv(ankle_data_path, index_col=index_col)

        self.support_info = utils.SupportPhaseInfo(self.trajectory_data["Actual_LeftFootZ"],
                                              self.trajectory_data["Actual_RightFootZ"],
                                              self.support_data["Target_SupportPhase"])
        self.collision_info = utils.CollisionInfo(self.collision_data["BallCollision_ForceAmplitude"].index,
                                                  self.collision_data["BallCollision_ForceAmplitude"])

circle = LogData("./unstabilized_logs", 1)
straight = LogData("./unstabilized_logs", 2)
straight_stabilized = LogData("./stabilized_logs", 2)
circle_stabilized = LogData("./stabilized_logs", 1)
disturbed_unstabilized = LogData("./disturbed_logs", 2)
disturbed_stabilized = LogData("./disturbed_logs", 1)
```

```python
T = 0.01
z = 0.86
g = 9.81
N = 100
stepwidth = 0.6

A = np.array([[1, T, T**2/2], [0, 1, T], [0, 0, 1]])
B = np.array([[T**3/6], [T**2/2], [T]])
C = np.array([[1, 0, -z/g]])

zmp = np.zeros(N)
zmp_ref = np.zeros(N)
zmp_ref[N/3:2*N/3] = stepwidth/2*np.ones(N-2*N/3)

x = np.zeros(3*N)
x.shape = (3, N)
u = np.zeros(N)

for i in range(1, N):
    zmp[i-1] = C.dot(x[:,[i-1]])
    u[i-1] = (zmp_ref[i] - zmp[i-1] - T * x[1, i-1] - T**2/2 * x[2, i-1]) / (T**3 / 6 - z/g * T + T**2)
    x[:, [i]] = A.dot(x[:, [i-1]]) + B * u[i-1]


fig, axes = plt.subplots(1, 2, figsize=(10, 2), sharex=True, sharey=False)
time = np.arange(0, T*N, T)
df_input = pd.DataFrame({'u': u, 'CoM Acceleration': 100*x[2,:]}, index=time)
df_input.plot(ax=axes[0])
plt.axes(axes[0])
plt.legend(["CoM Acceleration", "u"], loc=3, fontsize='small')
plt.xlabel("time / s")

df = pd.DataFrame({'ZMP Reference': zmp_ref, 'CoM': x[0,:], 'ZMP': zmp}, index=time)
df.plot(ax=axes[1])
plt.axes(axes[1])
plt.legend(["CoM", "ZMP", "Reference ZMP"], loc=3, fontsize='small')
plt.xlabel("time / s")

plt.savefig("../images/simple_zmp_control.png", bbox_inches="tight", pad_inches=0.2, dpi=300)
```
