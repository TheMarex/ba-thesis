```python
%pylab inline

import pandas as pd
import matplotlib.pyplot as plt

import utils

size = (18, 6)

# Finds newest log and calls pd.read_csv(..)
index_col = "Timestamp"

class LogData:
    def __init__(self, log_dir, n):
        com_data_path         = utils.get_nth_newest(log_dir, "com.csv", n)
        zmp_data_path         = utils.get_nth_newest(log_dir, "zmp.csv", n)
        trajectory_data_path  = utils.get_nth_newest(log_dir, "trajectory.csv", n)
        support_data_path     = utils.get_nth_newest(log_dir, "support.csv", n)
        collision_data_path   = utils.get_nth_newest(log_dir, "collision.csv", n)

        print("""
        Using the following logs:
        com:         %s
        zmp:         %s
        trajectory:  %s
        collision:   %s
        support:     %s
        """ % (com_data_path,
               zmp_data_path,
               trajectory_data_path,
               collision_data_path,
               support_data_path))

        self.com_data         = pd.read_csv(com_data_path, index_col=index_col)
        self.zmp_data         = pd.read_csv(zmp_data_path, index_col=index_col)
        self.trajectory_data  = pd.read_csv(trajectory_data_path, index_col=index_col)
        self.collision_data   = pd.read_csv(collision_data_path, index_col=index_col)
        self.support_data     = pd.read_csv(support_data_path, index_col=index_col)

        self.support_info = utils.SupportPhaseInfo(self.trajectory_data["Actual_LeftFootZ"],
                                              self.trajectory_data["Actual_RightFootZ"],
                                              self.support_data["Target_SupportPhase"])
        self.collision_info = utils.CollisionInfo(self.collision_data["BallCollision_ForceAmplitude"].index,
                                                  self.collision_data["BallCollision_ForceAmplitude"])

circle = LogData("./unstabilized_logs", 1)
straight = LogData("./unstabilized_logs", 2)
```

```python
# Plot for noisy CoM Acceleration

size=(10,2)
straight.com_data["Actual_CoMAccX"].plot(color='b', linewidth=0.5)
straight.com_data["Trajectory_CoMAccX"].plot(color='g', figsize=size, linewidth=1.5)
straight.support_info.add_support_phase_spans()
plt.ylim([-1500, 1500])
plt.xlim([4, 8])
plt.xlabel("time / s")
plt.ylabel("$\ddot{c}_x$ / mmm")
plt.legend(["Actual", "Desired"], loc=3, fontsize='small')
plt.savefig("../images/noisy_com_acc.png", bbox_inches="tight", pad_inches=0.2, dpi=300)
    
```

```python
# Plot ZMP comparison

straight.zmp_data["Trajectory_ZMPX"].plot(color='r')
straight.zmp_data["Actual_ZMPX"].plot(color='g', linewidth=0.5)
straight.zmp_data["MultiBody_ZMPX"].plot(color='b', linewidth=0.5, figsize=size)
plt.ylim([-200, 300])
straight.support_info.add_support_phase_spans()
plt.xlim([4, 8])
plt.xlabel("time / s")
plt.ylabel("x / mmm")
plt.legend(["Desired", "Cart-Table", "Multi Body"], loc=3, fontsize='small')
plt.savefig("../images/zmp_comparison.png", bbox_inches="tight", pad_inches=0.1, dpi=300)
```

```python
straight.zmp_data[["AngularMomentumX", "AngularMomentumY", "AngularMomentumZ"]].plot(figsize=size)
straight.support_info.add_support_phase_spans()
plt.xlim([2, 13])
plt.xlabel("time / s")
plt.ylabel("x / mmm")
plt.legend(["X", "Y", "Z"], loc=3, fontsize='small')
plt.savefig("../images/angular_momentum.png", bbox_inches="tight", pad_inches=0.1, dpi=300)
```

```python
# Plot state related to the state of stabilizer

target_trajectory_cols = [\
    "Target_LeftFootX", "Target_RightFootX", "Target_LeftFootY", "Target_RightFootY"
]
actual_trajectory_cols = [\
     "Actual_LeftFootX", "Actual_RightFootX", "Actual_LeftFootY", "Actual_RightFootY"
]
zmp_cols = [\
"Trajectory_ZMPX", "Trajectory_ZMPY", "Reference_ZMPX", "Reference_ZMPY","MultiBody_ZMPX", "MultiBody_ZMPY"
]
com_cols = [\
"Actual_CoMX", "Actual_CoMY", "Trajectory_CoMX", "Trajectory_CoMY"
]

zmp_controll_data = straight.zmp_data[zmp_cols]
has_trajectory_data = target_trajectory_cols[0] in straight.trajectory_data
if has_trajectory_data:
    zmp_controll_data[target_trajectory_cols] = straight.trajectory_data[target_trajectory_cols]
zmp_controll_data[actual_trajectory_cols] = straight.trajectory_data[actual_trajectory_cols]
zmp_controll_data[com_cols] = straight.com_data[com_cols]

size=(10,2)
fig, axes = plt.subplots(2, 1, figsize=(size[0], size[1]*2), sharex=True, sharey=True)
fig.set_dpi(300)
plt.axes(axes[0])
plt.xlim([2, 13])
plt.ylim([-100, 250])
plt.subplots_adjust(hspace = 0.05)
target_cols_x = ["Trajectory_CoMX", "Trajectory_ZMPX", "Reference_ZMPX"]
if has_trajectory_data:
    target_cols_x += ["Target_LeftFootX", "Target_RightFootX"]
zmp_controll_data[target_cols_x].plot(ax=axes[0])
plt.ylabel("x / mmm")
plt.xlabel("")
straight.support_info.add_support_phase_spans()
plt.legend(["CoM", "ZMP", "Reference ZMP"], loc=3, fontsize='small')

plt.axes(axes[1])
zmp_controll_data[["Actual_CoMX"]].plot(ax=axes[1], color='b')
zmp_controll_data[["MultiBody_ZMPX"]].plot(ax=axes[1], color='g', linewidth=0.5)
zmp_controll_data[["Actual_LeftFootX", "Actual_RightFootX"]].plot(ax=axes[1], color=('r', 'cyan'))
straight.support_info.add_support_phase_spans()
plt.ylabel("x / mmm")
plt.xlabel("time / s")
plt.legend(["CoM", "ZMP", "Right Foot", "Left Foot"], loc=3, fontsize='small')

plt.savefig("../images/player_undisturbed_straight_x.png", bbox_inches="tight", pad_inches=0.2, dpi=300)
    
fig, axes = plt.subplots(2, 1, figsize=(size[0], size[1]*3), sharex=True, sharey=True)
plt.axes(axes[0])
plt.xlim([2, 13])
plt.ylim([-100, 1500])
plt.subplots_adjust(hspace = 0.05)
target_cols_x = ["Trajectory_CoMY", "Trajectory_ZMPY", "Reference_ZMPY"]
if has_trajectory_data:
    target_cols_x += ["Target_LeftFootY", "Target_RightFootY"]
zmp_controll_data[target_cols_x].plot(ax=axes[0])
axes[0].set_ylabel("y / mmm")
axes[0].set_xlabel("")
straight.support_info.add_support_phase_spans()
plt.legend(["CoM", "ZMP", "Reference ZMP"], loc=3, fontsize='small')

plt.axes(axes[1])
zmp_controll_data[["Actual_CoMY", "MultiBody_ZMPY", "Actual_LeftFootY", "Actual_RightFootY"]].plot(ax=axes[1])
straight.support_info.add_support_phase_spans()
plt.xlabel("time / s")
plt.ylabel("y / mmm")
plt.legend(["CoM", "ZMP", "Right Foot", "Left Foot"], loc=3, fontsize='small')

plt.savefig("../images/player_undisturbed_straight_y.png", bbox_inches="tight", pad_inches=0)
    
```
